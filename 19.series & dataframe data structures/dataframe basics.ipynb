```python
df4 = pd.DataFrame([10,20,30,40], columns=['c1'], index=[1,3,5,7])
df4
```

```python
#slice the required rows using direct indexing
print(df4[0:2])
print(df4[1:3])
```

```python
#iloc is used to slice the data using implicit index
print(df4.iloc[0:2])
print(df4.iloc[1:3])
```

```python
#loc is used to slice the data using explicit index
```

```python
print(df4.loc[1:3])
print(df4.loc[:5])
```

```python
#using iloc on titanic data
print(titanic.iloc[0:10]) #row slicing
print(titanic.iloc[0:3, 0:3]) # row & column  slicing using slice operator
print(titanic.iloc[:10, :5])
print(titanic.iloc[[0,3],[0,2]]) # row & column slicing using fancy indexing
print(titanic.iloc[titanic.Sex=='male',[0,2]]) # row & column slicing using mask
```

```python
#using loc on titanic data
print(titanic.loc[0:10]) #row slicing
print(titanic.loc[0:3, 'Pclass':'Fare']) # row & column  slicing using slice operator
print(titanic.loc[:10, 'Pclass':'Fare'])
print(titanic.loc[[0,3],['Pclass','Sex','Fare']]) # row & column slicing using fancy indexing
print(titanic.loc[titanic.Sex=='male',['Pclass','Sex','Fare']]) # row & column slicing using mask
```
