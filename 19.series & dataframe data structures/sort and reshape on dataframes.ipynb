```python
import pandas as pd
```

```python
titanic = pd.read_csv('C:/users/algorithmica/python/titanic.csv')
titanic.info()
```

```python
titanic1 = titanic.sort_values('Age')
titanic1.head(10)
```

```python
#single column sort on dataframes
titanic2 = titanic.sort_values('Age', ascending=False)
titanic2.head(10)
```

```python
#multi column sort on dataframes
titanic3 = titanic.sort_values(['Age','Fare'], ascending=[False,True])
titanic3.head(10)
```
